# Removing punctuation marks and Stemming the Data

```python
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
    
    text = y[:]
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation and i not in y:
            y.append(i)
        if i not in stopwords.words('french') and i not in string.punctuation and i not in y:
            y.append(i);
        if i not in stopwords.words('german') and i not in string.punctuation and i not in y:
            y.append(i);
            
    text = y[:]
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
    
            
    return " ".join(y)
```

```python
df['transformed_text'] = df['text'].apply(transform_text)
```
